```python
import pandas as pd

data = {
    "name": ["Alex", "Bob", "Charles"],
    "age": [23, 34, 25]
}

df = pd.DataFrame(data, index = ['a', 'b', 'c'])
print(df)
```

```python
row = df.loc['b']
print(row)
```

```python
cell = df.loc['c', 'age']
print(cell)
```

```python
subset = df.loc[['a', 'b'], ['name', 'age']]
print(subset)
```

```python
filtered = df.loc[df['age'] > 23]
print(filtered)
```

```python
data = {
    "category": ['A', 'B', 'A', 'B', 'A', 'C'],
    'values': [10, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)
```

```python
grouped = df.groupby('category')
print(grouped.mean())
```

```python
data = {
    "category": ['A', 'A', 'B', 'B', 'C', 'C'],
    "sub-category": ["X", "Y", "X", "Y", "X", "Y"],
    'values': [10, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)
grouped = df.groupby(['category', 'sub-category']).sum()
print(grouped)
```

```python
cust_agg = df.groupby("category").agg({"values" : lambda x: x.max() - x.min()})
print(cust_agg)
```

```python
df1 = pd.DataFrame({'A': [1, 2, 3]}, index = ['a', 'b', 'c'])
df2 = pd.DataFrame({'B': [4, 5, 6]}, index = ['a', 'b', 'c'])
df3 = df1.join(df2, how="outer")
print(df3)
```

```python
df1 = pd.DataFrame({'key': ['a', 'b', 'c'], 'A': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'], 'B': [4, 5, 6]})
df3 = pd.merge(df1, df2, on = 'key', how = 'outer')
print(df3)
```

```python
df1 = pd.DataFrame({'A': [1, 2]}, index = ['a', 'b'])
df2 = pd.DataFrame({'B': [2, 4]}, index = ['b', 'c'])
df3 = pd.concat([df1, df2], axis = 1)
print(df3)
```
